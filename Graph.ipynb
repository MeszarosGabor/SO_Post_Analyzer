```python
import networkx as nx
```

```python
g = nx.Graph()
g.add_nodes_from(['os', 'sys', 'pathlib'])
g.add_edges_from([('os', 'pathlib'), ('os', 'sys')])
g.add_nodes_from(['numpy', 'pandas', 'sklearn'])
g.add_edges_from([('pandas', 'numpy'), ('pandas', 'sklearn'), ('numpy', 'sklearn'), ('numpy', 'os')])
pos = nx.spring_layout(g, k=100.0)

color_map = {
    #"os":
    "sys": 'lime',
    "pathlib": "yellow", 
    "numpy": "lime",
    "pandas": "aqua",
    "sklearn": "aqua",
}

nx.draw(g, with_labels=True,
        node_color = [color_map.get(node, "grey") for node in g.nodes()],
       node_size=1000, linewidths=1)
```
